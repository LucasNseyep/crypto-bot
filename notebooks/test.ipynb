```python
import ccxt
import pandas as pd

# 1) make an exchange
exchange = ccxt.binance({
    "enableRateLimit": True
})

# 2) pick a market + timeframe
symbol = "BTC/USDT"
timeframe = "1h"   # you can try "5m" or "1m" too
limit = 100        # how many candles to fetch

# 3) fetch ohlcv: [ timestamp, open, high, low, close, volume ]
ohlcv = exchange.fetch_ohlcv(symbol, timeframe=timeframe, limit=limit)

# 4) put it in a DataFrame so it's nice to look at
df = pd.DataFrame(
    ohlcv,
    columns=["timestamp", "open", "high", "low", "close", "volume"]
)

# 5) turn timestamp into readable datetime (UTC)
df["datetime"] = pd.to_datetime(df["timestamp"], unit="ms")

# 6) show it
print(df.tail(10))
```
